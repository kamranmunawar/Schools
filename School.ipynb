```python
import os
folder_files = os.listdir(r'postcodes') 
print("This Folder contains {len_folder} file(s).".format(len_folder=len(folder_files)))
filepath = (r'postcodes')
for file in folder_files:
    print(os.path.join(filepath, file))
    #print(file_name)
    file_name = os.path.join(filepath, file)
    postcodefile = pd.read_csv(file_name)
    poscodes = postcodefile.loc[postcodefile['Postcode'].isin(schoolinfo.index)]
    schoolinfo.update(poscodes.set_index('Postcode'))
    #print(filename)
```
